```python
path = f'{os.getcwd()}\\Garp vs Aokiji.mp4'
video = cv.VideoCapture(path)
frames = []
```

```python
def high_contrast(img):
    img[img < 50] = 0
    img[(img >= 50) & (img < 125)] = 75
    img[(img >= 125) & (img < 200)] = 175
    img[img >= 200] = 255
    return img
```

```python
while True:
    ret, frame = video.read()
    if ret == False:
        print("the video is finished")
        break
    gray = cv.cvtColor(frame , cv.COLOR_BGR2GRAY)
    img = Image.fromarray(gray)
    fram = img.resize((100, 100), resample=Image.Resampling.LANCZOS)
    frames.append(fram)
    #cv.imshow('Garp', img)
    #cv.resizeWindow('Garp', 600, 400)
    '''x += 1
    if x == 1000:
        break'''
    #if cv.waitKey(1) == ord('q'):
     #  break
```

```python
frames = np.array(frames)
frames.shape
```

```python
def img2ascii(img):# 75, 100, 125, 150, 175, 200, 255 
    pic = high_contrast(img).astype(str)
    pic[pic == '255'] = "  "
    pic[pic == '175'] = "*"
    pic[pic == '75'] = "#"
    pic[pic == '0'] = "@"
    pic[pic  == '0.0'] = "@"
    #output = "\n".join(",".join(row) for row in pic)
    return pic.astype('U1') #output.replace(',','')
```

```python
def ascii_frame(frames):
    ascii_frame = []
    for img in frames:
        pic = img2ascii(img)
        ascii_frame.append(pic)
    return ascii_frame
```

```python
def gen_frame(img):
    simp = np.full((1155, 1155), 255, dtype = np.uint8)
    pos = 14
    for i in img:
        f_size = 1
        co = (0 ,0 ,0)
        f_family = cv.FONT_HERSHEY_COMPLEX_SMALL
        thic = 2
        text = ''.join(i)
        pic = cv.putText(simp, text, (1,pos), f_family, 1, co, thic)
        pos += 14    
    return pic
```

```python
ascii_frames = ascii_frame(frames)
```

```python
video = []
for idx,i in enumerate(ascii_frames):
    print(f'------------------------- the {idx + 1} Frame -----------------------------')
    vid = gen_frame(i)
    video.append(vid)
```

```python
output = cv.VideoWriter('gerp.mp4', cv.VideoWriter_fourcc(*'mp4v'), 30,(1727, 1727), False)
```

```python
for frame in video:
    output.write(frame)
output.release()
```
